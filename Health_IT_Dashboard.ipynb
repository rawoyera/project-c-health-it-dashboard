```python
import pandas as pd
from google.colab import files

# Upload file
uploaded = files.upload()

# Load into dataframe
df = pd.read_excel('Fitness_Health_Tracking_Dataset_with_Missing_Values.xlsx')

# Fix Workout_Type nulls
df['Workout_Type'] = df['Workout_Type'].fillna('Rest/None')

# Fix numeric nulls per user
numeric_cols = ['Hours_Slept', 'Water_Intake (Liters)', 'Active_Minutes', 'Heart_Rate (bpm)']
for col in numeric_cols:
    df[col] = df.groupby('User_ID')[col].transform(lambda x: x.fillna(x.median()))

print("Done! Null check:")
print(df.isnull().sum())
```

```python
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Add Month and Quarter columns
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter

print(df[['Date', 'Month', 'Quarter']].head())
```

```python
df.to_csv('cleaned_fitness_data.csv', index=False)
print("Exported! Shape:", df.shape)
```

```python
import pandas as pd

# Load your clean file
df = pd.read_csv('cleaned_fitness_data.csv')

# ── EXPORT 2: Monthly Trends ──────────────────────────────────────────
# Group by Month, take the average of each key metric
monthly_trends = df.groupby('Month').agg(
    Avg_Steps=('Steps_Taken', 'mean'),
    Avg_Sleep=('Hours_Slept', 'mean'),
    Avg_Heart_Rate=('Heart_Rate (bpm)', 'mean'),
    Avg_Stress=('Stress_Level (1-10)', 'mean'),
    Avg_Active_Minutes=('Active_Minutes', 'mean'),
    Avg_Calories=('Calories_Burned', 'mean')
).reset_index()

# Round to 2 decimal places — cleaner in dashboard labels
monthly_trends = monthly_trends.round(2)

monthly_trends.to_csv('monthly_trends.csv', index=False)
print("✅ monthly_trends.csv exported")
print(monthly_trends)

# ── EXPORT 3: KPI Summary ─────────────────────────────────────────────
# WHO recommends ≥150 active minutes/week → ≥21.4 min/day average
# We'll flag each row, then calculate what % of user-days meet the threshold

df['Meets_Activity_Guidelines'] = df['Active_Minutes'] >= 21.4

kpi_summary = pd.DataFrame([{
    'Avg_Daily_Steps': round(df['Steps_Taken'].mean(), 0),
    'Avg_Hours_Slept': round(df['Hours_Slept'].mean(), 2),
    'Avg_Heart_Rate_bpm': round(df['Heart_Rate (bpm)'].mean(), 1),
    'Avg_Stress_Level': round(df['Stress_Level (1-10)'].mean(), 2),
    'Pct_Meeting_Activity_Guidelines': round(
        df['Meets_Activity_Guidelines'].mean() * 100, 1
    )
}])

kpi_summary.to_csv('kpi_summary.csv', index=False)
print("\n✅ kpi_summary.csv exported")
print(kpi_summary)
```
